```python
data = data.merge(country_group, on = 'native-country', how='left')
```

```python
data
```

```python
data.drop('native-country', axis=1,inplace=True)
data = data.rename(columns={'class_x':'class', 'class_y':'native-country'})
```

```python
data
```

```python
data.isna().mean() # 데이터 전처리 결측치 처리및 더미 변수 변환

# 3개의 변수에서 결측치 workclass, occupation, native-country 어떤 값을 채워야 적당할까?
```

```python
data['native-country'] = data['native-country'].fillna(-99) #결측치를 -99로 대체
```

```python
data['workclass'].value_counts()
```

```python
data['workclass'] = data['workclass'].fillna('Private')
```

```python
data['occupation'].value_counts()
```

```python
data['occupation'] = data['occupation'].fillna('Unknown')
```

```python
data = pd.get_dummies(data, drop_first=True) # 모든 결칙치 해결 했으니 더미 변수로 변환
```

```python
from sklearn.model_selection import train_test_split
X = data.drop('class',axis=1)
y = data['class']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.4, random_state=100)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
pred= model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred= model.predict(X_test)
print('Train score:', accuracy_score(y_train, train_pred), 'Test score:',
      accuracy_score(y_test,test_pred)) # 훈련셋에서는 정확도가 높지만, 트레이닝셋은 정확도가 낮음 so 오버피팅
```

```python
model = DecisionTreeClassifier(max_depth=7)
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred= model.predict(X_test)
print('Train score:', accuracy_score(y_train, train_pred), 'Test score:',
      accuracy_score(y_test,test_pred))
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(30,15))
plot_tree(model, max_depth=3,fontsize=15,
         feature_names=X_train.columns) #max depth =3 으로 제한해서 중요하게 여겨지는 3단계 까지만 확인 
plt.show()
```
