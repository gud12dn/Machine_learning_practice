```python
data = data.dropna(subset=[
    'pref_o_attractive', 'pref_o_sincere', 'pref_o_intelligence', 
    'pref_o_funny', 'pref_o_ambitious', 'pref_o_shared_interests', 
    'attractive_important', 'sincere_important', 'intellicence_important', 
    'funny_important', 'ambtition_important', 'shared_interests_important'
])  # 일부 변수에서 결측치 제거
```

```python
data = data.fillna(-99)
# -99와 같은 숫를 넣어서 해당 사람은 해당 항목에 응답하지 않음을 나타냄 
```

```python
data.isna().mean()
```

```python
def age_gap(x):
    if x['age'] == -99:
        return -99
    elif x['age_o'] == -99:
        return -99
    elif x['gender'] == 'female':
        return x['age_o'] - x['age']
    else:
        return x['age'] - x['age_o']

# 함수 생성 남녀중 한명이라도 나이가 -99 dlaus -99 반환 그렇지 않으면 남자가 연상이면 플러스 값이, 여자가 연상이면 마이너스 값 반환
```

```python
data['age_gap'] = data.apply(age_gap,axis=1)
```

```python
data['age_gap_abs'] = abs(data['age_gap']) # 나이 자체가 중요한 변수가 될 수 있게 absolute 절댓값 만들기
```

```python
def same_race(x):
    if x['race'] == -99:
        return -99
    elif x['race_o'] == -99:
        return -99
    elif x['race'] == 'race_o':
        return 1
    else:
        return -1
```

```python
data['same_race'] = data.apply(same_race, axis=1)
# data를 same_race 함수에 적용하여 결과를 same_race 변수로 저장
```

```python
def same_race_point(x):
    if x['same_race'] == -99:
        return -99
    else:
        return x['same_race'] * x['importance_same_race']
# same_race와 importance_same_race의 곱을 리턴
```

```python
data['same_race_point'] = data.apply(same_race_point, axis=1)

# 인종 여부 중요도 파악 하기 위해 이 작업을 진행함
```

```python
def rating(data, importance, score):
    if data[importance] == -99:
        return -99
    elif data[score] == -99:
        return -99
    else:
        return data[importance] * data[score]

# importance 와 score의 곱을 리턴
```

```python
#총 4개의 범주의 변수 이름 리스트 만들기 

partner_imp = data.columns[8:14] # 상대방의 중요도
partner_rate_me = data.columns[14:20] # 본인에 대한 상대방의 평가
my_imp = data.columns[20:26] # 본인의 중요도
my_rate_partner = data.columns[26:32] # 상대방에 대한 본인의 평가
```

```python
# 상대방 관련 새 변수 이름을 저장하는 리스트
new_label_partner = ['attractive_p', 'sincere_partner_p', 'intelligence_p', 
                     'funny_p', 'ambition_p', 'shared_interests_p']

# 본인 관련 새 변수 이름을 저장하는 리스트
new_label_me = ['attractive_m', 'sincere_partner_m', 'intelligence_m', 
                'funny_m', 'ambition_m', 'shared_interests_m']

# 평가 점수 x 중요도을 받아줄 새 변수의 이름 만들어줌 
```

```python
for i,j,k in zip(new_label_partner, partner_imp, partner_rate_me):
    print(i,' & ',j,' & ', k)

# 위의 6개의 리스트를 한 번에 3가지(새 변수 이름, 중요도, 평가) 리스트를 사용해 계산
```

```python
for i, j, k in zip(new_label_partner, partner_imp, partner_rate_me):  # 순회
    data[i] = data.apply(lambda x: rating(x, j, k), axis=1)
    # rating 함수 적용 결과를 i 변수에 저장
```

```python
for i, j, k in zip(new_label_me, my_imp, my_rate_partner):  # 순회
    data[i] = data.apply(lambda x: rating(x, j, k), axis=1)
    # rating 함수 적용 결과를 i 변수에 저장
```

```python
data = pd.get_dummies(data, columns=['gender', 'race', 'race_o'], drop_first=True)
# 더미 변수로 변환 object 타입 변수들은 문자 형태이기 때문에 숫자 형태가 되게끔 더미 변수로 바꿔줍니다
```
