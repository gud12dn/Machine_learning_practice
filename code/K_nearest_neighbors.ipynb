```python
mm_scaler = MinMaxScaler()
mm_scaled = mm_scaler.fit_transform(data)
mm_scaled = pd.DataFrame(mm_scaled, columns= data.columns)
round(mm_scaled.describe(),2)
```

```python
from sklearn.model_selection import train_test_split
X = data.drop('class',axis=1)
y = data['class']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.2, random_state=100)
```

```python
mm_scaler = MinMaxScaler()
mm_scaler.fit(X_train)
```

```python
X_train_scaled = mm_scaler.transform(X_train)
X_test_scaled = mm_scaler.transform(X_test)
```

```python
mm_scaler = MinMaxScaler()
X_train_scaled = mm_scaler.fit_transform(X_train)
X_test_scaled = mm_scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train_scaled,y_train)
pred = knn.predict(X_test_scaled)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
knn = KNeighborsClassifier(n_neighbors=7) # KNN 모델 생성 파라미터를 기본값 5 에서 7로 바꿈
knn.fit(X_train_scaled,y_train)
pred = knn.predict(X_test_scaled)
accuracy_score(y_test, pred) # 92%로 높아진것을 볼 수 있다
```

```python
for i in range(1,21):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_scaled,y_train)
    pred = knn.predict(X_test_scaled)
    print(accuracy_score(y_test, pred))
    
    
```

```python
scores = []
```

```python
for i in range(1,21):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_scaled,y_train)
    pred = knn.predict(X_test_scaled)
    acc = accuracy_score(y_test, pred)
    scores.append(acc)
```

```python
scores
```

```python
sns.lineplot(x=range(1,21), y=scores)
```

```python
knn = KNeighborsClassifier(n_neighbors=13) #13이 가장 최적의 하이퍼 파라미터
knn.fit(X_train_scaled,y_train)
pred = knn.predict(X_test_scaled)
accuracy_score(y_test, pred) 
```
